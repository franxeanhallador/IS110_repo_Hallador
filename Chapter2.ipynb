CHALLENGE

Write a program which prompts the user for a Celsius temperature, convert the temperature to Fahrenheit, and print out the converted temperature.

[PEW] E2.5

````python
# 1. Prompt the user for a Celsius temperature
celsius_str = input("Enter temperature in Celsius: ")

# 2. Convert the input string to a number (using float for accuracy)
celsius = float(celsius_str)

# 3. Calculate the Fahrenheit equivalent
fahrenheit = (celsius * 9/5) + 32

# 4. Print the result
print(f"{celsius} degrees Celsius is {fahrenheit} degrees Fahrenheit")
````

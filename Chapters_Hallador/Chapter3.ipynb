Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:



```
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

[PEW] E3.2

````python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    reg_pay = 40 * rate
    overtime_hrs = hours - 40
    ot_pay = overtime_hrs * (rate * 1.5)
    total_pay = reg_pay + ot_pay
else:
    total_pay = hours * rate

print(f"Pay: {total_pay}")
````

# CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:



```
Score           Grade
[ 98, 100]      1.0
[ 94,  97)      1.25
[ 90,  93)      1.5
[ 86,  89)      1.75
[ 82,  85)      2.0
[ 78,  81)      2.25
[ 75,  77)      2.5
[ 0,   74)      5.0
```

Adapted from [PEW] E3.3

````python
def computegrade(score):
    if score >= 98:
        return "1.0"
    elif score >= 94:
        return "1.25"
    elif score >= 90:
        return "1.5"
    elif score >= 86:
        return "1.75"
    elif score >= 82:
        return "2.0"
    elif score >= 78:
        return "2.25"
    elif score >= 75:
        return "2.5"
    else:
        return "5.0"

score_input = input("Enter Score: ")
score = float(score_input)

if score < 0 or score > 100:
    print("Error: Score must be between 0 and 100.")
else:
    print(computegrade(score))
````

Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).



```
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

[PEW] E4.6

````python
def computepay(hours, rate):

    if hours > 40:
        reg_pay = 40 * rate
        ot_hrs = hours - 40
        ot_pay = ot_hrs * (rate * 1.5)
        pay = reg_pay + ot_pay
    else:

        pay = hours * rate

    return pay

h_str = input("Enter Hours: ")
r_str = input("Enter Rate: ")


if h_str.replace('.', '', 1).isdigit() and r_str.replace('.', '', 1).isdigit():

    h = float(h_str)
    r = float(r_str)

    total_pay = computepay(h, r)

    print(f"Pay: {total_pay}")
else:
    print("Error: Please enter numeric values.")
````

# CHALLENGE

Challenge

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:



```
Score           Grade
[ 98, 100]      1.0
[ 94,  97)      1.25
[ 90,  93)      1.5
[ 86,  89)      1.75
[ 82,  85)      2.0
[ 78,  81)      2.25
[ 75,  77)      2.5
[ 0,   74)      5.0

Run the program repeatedly to test the various different values for input.
```

Adapted from [PEW] E4.6

````python
def computegrade(score):
    if score >= 98:
        return "1.0"
    elif score >= 94:
        return "1.25"
    elif score >= 90:
        return "1.5"
    elif score >= 86:
        return "1.75"
    elif score >= 82:
        return "2.0"
    elif score >= 78:
        return "2.25"
    elif score >= 75:
        return "2.5"
    else:
        return "5.0"

print("Grade Calculator (Type 'exit' to quit)")

while True:
    s_str = input("Enter Score (0-100): ")

    if s_str.lower() == 'exit':
        break

    if s_str.replace('.', '', 1).isdigit():
        s = float(s_str)

        if 0 <= s <= 100:
            grade = computegrade(s)
            print(f"Grade: {grade}")
        else:
            print("Error: Score must be between 0 and 100.")
    else:
        print("Error: Please enter a numeric value.")
````
